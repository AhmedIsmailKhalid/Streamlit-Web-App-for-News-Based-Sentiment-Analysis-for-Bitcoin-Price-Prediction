Having explored the distribution of words and their sentiments as well as the distribution for the articles and their sentiments, now lets perform Step 1 of our Statistical Driven Approach for calculating the sentiment scores for the words that are not in the Vader Lexicon

```python
news_df['temp_list'] = news_df['Text'].apply(lambda x:str(x).split())
```

```python
# Performing Step 1
words = []
for row in news_df['temp_list']:
    words = words + row
```

```python
words = np.unique(np.array(words))#.tolist()
```

```python
words = [re.split('(\d+)', word) for word in words]
words[:20]
```

```python
stop_words.append('k')
stop_words.append('s')
stop_words.append('x')
stop_words.append('m')
stop_words.append('')
```

```python
words_ = []
for sublist in words:
    for word in sublist :
        if word not in stop_words:
            words_.append(word)
```

```python
# In the interest of saving memory, let's update words and delete words_
words = words_
del(words_)
```

```python
words = [word for word in words if len(word) > 3]
```

```python
Analyzer = SentimentIntensityAnalyzer()

news_df['Compound Score'] = [Analyzer.polarity_scores(x)['compound'] for x in news_df['Text']]
news_df['Positive Score'] = [Analyzer.polarity_scores(x)['pos'] for x in news_df['Text']]
news_df['Neutral Score'] = [Analyzer.polarity_scores(x)['neu'] for x in news_df['Text']]
news_df['Negative Score'] = [Analyzer.polarity_scores(x)['neg'] for x in news_df['Text']]
news_df
```

```python
btc_df = btc_df[btc_df['Date'].isin(news_df['Date'])]
btc_df = btc_df.sort_values('Date',ascending=False).reset_index(drop=True)
btc_df
```

```python
filtered_news_counts = news_df.groupby(['Date','Time']).count()
filtered_news_counts = filtered_news_counts.sort_values('Time',ascending=False).reset_index()
filtered_news_counts.head()
```

```python
lst = []
for row,count in zip(btc_df.iterrows(),filtered_news_counts['Text']) :
    for i in range(count) :
        lst.append(row[1])
        

btc_df2 = pd.DataFrame(lst)
btc_df2 = btc_df2.reset_index(drop=True)
btc_df2
```

```python
scores = []
for value in btc_df2[['Close','Open']].iterrows() :
    score = (value[1][0] - value[1][1]) / btc_df2['Close'].max()
    scores.append(score)
```

```python
news_df['Fluctuation Scores'] = scores
news_df = news_df.fillna(0)
news_df.head()
```

```python
batches = []

for batch in batch_generator(words, 10):
    batches.append(batch)
```

```python
scores_dict = {}
WORDS = 0
for word in words :
    word_scores = []
    #print('\t\t\t\t',word, '\n')
    #print(word)
    WORDS += 1 
    for batch in batches:
        #print(Text)
        if word in batch :
            index = news_df.index[news_df['Text'].str.contains(word)]
            date = news_df.iloc[index]['Time']
            score = news_df.iloc[index]['Fluctuation Scores']
            word_scores.append(score)
        #if word not in Text :
        #    print('FOUND : No')
    word_score = np.mean(word_scores)
    #print(word, word_score)
    scores_dict[word] = word_score
    #print('***********************************************')
    if WORDS%1000 == 0 :
        print(str(WORDS) + '/' + str(len(words)) + ' words processed')
```

```python
scores_dict_df = pd.DataFrame(data = {'Word' : list(scores_dict.keys()), 'Score' : list(scores_dict.values())})
#scores_dict_df.to_csv('Semantic Scores Dict.csv', index=False)
scores_dict_df# = pd.read_csv('Semantic Scores Dict.csv')
```

```python
scaled_scores = []
for s in scores_dict_df['Score'] :
    scaled_score = (s - (-4)) / (4 - (-4))
    scaled_scores.append(scaled_score)
```

```python
scores_dict_df['Scaled Scores'] = scaled_scores
scores_dict_df.head()
```

```python
analyzer = SentimentIntensityAnalyzer()
for word in scores_dict_df['Word'] :
    try :    
        analyzer.lexicon.pop(word)
    except KeyError :
        continue
```
